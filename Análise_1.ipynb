importando biblioteca

```python
import pandas as pd
import numpy as np
```

Definindo DataFrame com arquivo csv

Utilizando apenas colunas solicitadas




```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deborah/Análises/Análise_1/Arquivos/State of Data 2021 - Dataset - Pgina1.csv', sep = ',',  usecols=[0,1,3,5,6,9,16,17,18])
df
```

Renomeando Colunas

```python
df.rename(columns={
    df.columns[0]: 'id',
    df.columns[1]: 'idade_antigo',
    df.columns[2]: 'genero',
    df.columns[3]: 'uf',
    df.columns[4]: 'regiao',
    df.columns[5]: 'nivel_ensino',
    df.columns[6]: 'cargo_atual',
    df.columns[7]: 'nivel',
    df.columns[8]: 'faixa_salarial'  
      }, inplace = True)

df
```

Criando nova coluna idade para verificar e tratar valores NaN

```python
df.loc[:, 'idade'] = df['idade_antigo']
df
```

Convertendo NaN para 0

```python
df.fillna({'idade': 0}, inplace = True)
df
```

Convertendo tipo de dados idade ( de float para inteiro )

```python
df['idade'] = df['idade'].astype(int)
df
```

Verificando valores menores que 18 incluindo 0

```python
df.idade[df['idade'] < 18]
```

Verificando dados menores que 18 em relação ao df

```python
(df.idade[df['idade'] < 18].count() / df.idade[df['idade']].count()) * 100
```

Verificando média coluna idade_antigo

```python
df['idade_antigo'].mean()
```

Média coluna idade, sem casa decimal e convertida a inteiro

```python
df['idade'].mean().round().astype(int)
```

Alterando valores idade = 0 para média de idade

```python
df.loc[df.idade == 0,'idade']= df['idade'].mean().round().astype(int)
df
```

Conferindo tratamento dados idade = 0 para média coluna idade

```python
df[df['idade_antigo'].isna()]
```

Salvando novo arquivo csv

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Deborah/Análises/Análise_1/Arquivos/Dataset1.csv', sep = ';')
```
